```python
from nitriding import NitridingLayer
from nitriding import equilibrate
import numpy as np
```

```python
equilibrate(943, 0.001)
```

```python
# Number of space cells.
nx = 200

# Thickness of system [m]
x_end = 0.0003

# Integration end time [s].
t_end = 60.0

# Set initial composition profile.
u0 = np.ones(nx) * 0.000100

# Reactive annealing condition.
# u0[0] = u0[-1] = 0.001000

# Fully austenitic surface
u0[0] = u0[-1] = 0.020000


# User defined function of time for temperature.
T = lambda t: 943.0

# Create model instance.
model = NitridingLayer()
```

```python
%%time
fig = model.solve_implicit(t_end, x_end, u0.copy(), T, nt=int(t_end)+1)
```
