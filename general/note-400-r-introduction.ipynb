## Using packages

````r
# There are two basic ways of loading a package, `require` which is
# the verbose version and `library`, as we have already discussed.
require(pacman)
library(pacman)
````

````r
# It is also possible to access a packages contents without loading
# by using the `::` access. For the previous iris dataset:
summary(datasets::iris)
````

````r
# Using `pacman` you can load multiple packages that way, including
# itself. This is not recommended for *base* packages.
pacman::p_load(pacman, dplyr, tidyr, ggplot2)
````

````r
# It also allow for unloading specific packages or all at once.
p_unload(dplyr, tidyr)
p_unload(all)
````

## Basic plotting

````r
# Reload the datasets package.
library(datasets)

# Reminder of the dataset head.
head(iris)
````

````r
# Categorical variable: automatic bar-plot.
plot(iris$Species)
````

````r
# Quantitative variable: index-based scatter-plot.
plot(iris$Petal.Length)
````

````r
# Categorical vs. quantitative: box-plots.
plot(iris$Species, iris$Petal.Width)
````

````r
# Pair of quantitative: scatter-plot.
plot(iris$Petal.Length, iris$Petal.Width)
````

````r
# Entire data frame: already shown...
# plot(iris)
````

````r
# If you actually intend to save for reporting, plot with options
# is a better choice for readability. Check below for some of the
# available options.
plot(iris$Petal.Length, iris$Petal.Width,
     col = "#cc0000",  # Hex code for red
     pch = 19,         # Use solid circles for points
     main = "Iris: Petal Length vs. Petal Width",
     xlab = "Petal Length",
     ylab = "Petal Width")
````

````r
# It is also possible to provide start and end points to plot
# provided functions. Next plot `cos`, `exp`, and `dnorm`.
plot(cos, 0, 2*pi)
plot(exp, 1, 5)
````

````r
plot(dnorm, -3, +3)
# ...activate a grid for last plot.
grid()
````

````r
# ... or add options for a functional plot.
plot(dnorm, -3, +3,
     col = "#cc0000",
     lwd = 5,
     main = "Standard Normal Distribution",
     xlab = "z-scores",
     ylab = "Density")
grid()
````

## Bar-charts and histograms

````r
# For this let's use another dataset. Check its description and
# first lines before plotting.
?mtcars
head(mtcars)
````

````r
# If you try as follows all is gonna be produced is a bar per row
# of the dataset, what is not very useful (or not useful at all).
barplot(mtcars$cyl)
````

````r
# To produce the frequency of each category use `table`
# Need a table with frequencies for each category.
cylinders <- table(mtcars$cyl)
barplot(cylinders)
````

````r
# Clear environment (all variables).
rm(list = ls()) 
````

````r
# Continuous variables cannot use the previous approach and a way
# to achieve similar results is by the use of histograms. Let's get
# back to iris dataset and check some of these:
head(iris)
hist(iris$Sepal.Length)
hist(iris$Sepal.Width)
hist(iris$Petal.Length)
hist(iris$Petal.Width)
````

````r
# Using unique, determine the number of species.
unique(iris$Species)
````

````r
# So let's compare the histograms of a given quantity in this set
# grouped by species and put all in same plot. For doing so we need
# to change the parameter `mfrow` to accommodate 3 sub-plots:
par(mfcol = c(1, 3), pin = c(2, 2))

# Histograms for each species using options. In order to ensure the
# plots are aligned, use the same `xlim` for all of them.
hist(iris$Petal.Width [iris$Species == "setosa"],
     xlim = c(0, 3),
     breaks = 9,
     main = "Petal Width for Setosa",
     xlab = "",
     col = "red")

hist(iris$Petal.Width [iris$Species == "versicolor"],
     xlim = c(0, 3),
     breaks = 9,
     main = "Petal Width for Versicolor",
     xlab = "",
     col = "purple")

hist(iris$Petal.Width [iris$Species == "virginica"],
     xlim = c(0, 3),
     breaks = 9,
     main = "Petal Width for Virginica",
     xlab = "",
     col = "blue")
````

````r
# Restore graphic parameter with single plot.
par(mfrow=c(1, 1))
````

````r
# To finish this first practical tutorial, let's check another set
# with annual Canadian Lynx trappings 1821-1934:
?lynx
head(lynx)
````

````r
# Check the very poor default histogram.
hist(lynx)
````

````r
# It can be improved with some options.
hist(lynx,
     breaks = 28,          # "Suggests" 28 bins
     freq   = FALSE,       # Axis shows density, not freq.
     col    = "thistle1",  # Color for histogram
     main   = paste("Histogram of Annual Canadian Lynx",
                    "Trappings, 1821-1934"),
     xlab   = "Number of Lynx Trapped")

# With `curve` we can add a normal distribution overlay to the
# previous plot. Make sure to set `add = TRUE` for doing so.
curve(dnorm(x, mean = mean(lynx), sd = sd(lynx)),
      col = "thistle4",  # Color of curve
      lwd = 2,           # Line width of 2 pixels
      add = TRUE)        # Superimpose on previous graph

# Add two kernel `density` estimators.
lines(density(lynx, adjust = 0.5), col = "blue", lwd = 2)
lines(density(lynx, adjust = 2.0), col = "purple", lwd = 2)

# Finish with a rug plot for showing actual positions inside bins.
rug(lynx, lwd = 2, col = "gray")
````
