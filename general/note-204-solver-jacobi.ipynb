# Jacobi iterative solver

In this notebook we explore the implementation of a Jacobi iteration solver applied to a simple case of Poisson equation. It solves the system $Au = f(x)$, where $A$ is the negative of a second difference matrix with three main diagonals occupied by $[-1 2 -1]$ and $f(x)$ is a function of position. The problem is solved for the *fixed-fixed* case, where both endings are held constant and using constant right-hand side $f(x)=1{}$.

We start by importing the required packages. All the numerical implementation will be done with `numpy` and visualization is done with classical `matplotlib`.

```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

This is a 1-D problem for which the lenght `L` is discretized into `size` nodes separated by a distance `dx`.  In the problem we make use of the squared value of `dx` to keep discretization scalable, i.e. make it tend toward the same solution disregarding the number of nodes.  Otherwise the problem solution would have different solution for different `size`.

```python
def get_discretization(L, size):
    """ Returns space discrete nodes for solution. """
    x, dx = np.linspace(0, L, size, retstep=True)
    return x, pow(dx, 2)
```

The analytical solution to this problem given the fixed boundaries is provided by the following polynomial. Compute a high-resolution `hr` version of solution for display and a low-resolution with the same grid as the problem for error computation.

```python
def get_analytical_solution(x, L, size):
    """ Returns problem analytical solution. """
    solution = lambda x: x * (1 - x) / 2
    hrx = np.linspace(0, L, 10 * size)
    hrua, lrua = solution(hrx), solution(x)
    return hrx, hrua, lrua
```

The residuals are generally evaluated as the norm of $f(x) - Au'$, where $u'$ is the approximate solution after a certain iteration.  Here we provide the matrix $A$ for computation of this quantity during solution. We do not apply boundary conditions to $A$ since they will remain unused. Also check the eigenvalues of $D^{-1}LU$ to ensure the system is convergent. The maximum allowed absolute value is unity.

```python
def get_problem_matrix(size, dx2, relax):
    """ Returns problem discrete representation. """
    L, D = -1.0 * np.ones(size-1), 2.0 * np.ones(size-0)
    A = (np.diag(L, -1) + np.diag(D, 0) + np.diag(L, 1)) / dx2
    
    R = (np.diag(L,-1) + np.diag(L,1)) / D
    w, v = np.linalg.eig(R)
    gamma = relax * max(np.abs(w)) + (1.0 - relax) 
    
    print("Maximum eigenvalue is {:.6f}".format(gamma))
    
    return A
```

To make the code more readable and well formatted, create slices for each of the ranges necessary in the computation. Once we are dealing with fixed
boundaries it is useful to declare the `inner` region of the system. The secondary diagonals that operate in Lu and Uu give rise to the `lower` and `upper` slices that make a central difference.

```python
INNER = slice(1, -1)
LOWER = slice(0, -2)
UPPER = slice(2, None)
```

```python
def jacobi(A, b, u, r, dx2, guess):
    advance = guess[LOWER] + guess[UPPER]
    u[INNER] = r * 0.5 * (dx2 * b[INNER] + advance) + (1 - r) * u[INNER]

    delta = np.max(np.abs(u[:] - guess[:]))
    resit = np.max(b[INNER] - A[INNER, INNER] @ guess[INNER])
    
    guess[INNER] = u[INNER]
    
    return guess, u, delta, resit
```

Since after solving the proble we will need to visualize results, we conceive `plot_solution` in advance so that it can be used later in the solver function. Here we display the solution in space, the residuals for each node (local error), the maximum update step per iteration and the convergence history. 

```python
def plot_solution(x, un, hrx, hrua, lrua, deltaerr, residual, counter):
    """ Plot solution and history of residuals. """
    title = F'{len(x)} points, converged after {counter} iterations'
    colors = plt.style.library['seaborn']['axes.prop_cycle']
    colors = colors.by_key()['color']
    
    plt.close('all')
    plt.style.use('seaborn-white')

    fig, ax = plt.subplots(2, 2, figsize=(12, 8))
    fig.suptitle(title)

    ax[0, 0].plot(x, un, 'o', ms=3, label='Numerical')
    ax[0, 0].plot(hrx, hrua, ':', lw=1, label='Analytical')
    ax[0, 0].grid(linestyle=':')
    ax[0, 0].set_xlabel('x')
    ax[0, 0].set_ylabel('u')
    ax[0, 0].legend()

    ax[0, 1].plot(x, lrua - un, 'o', ms=3, label='Local error')
    ax[0, 1].plot(x, lrua - un, ':', lw=1, label='_nolegend_',
                  c=colors[0])
    ax[0, 1].grid(linestyle=':')
    ax[0, 1].set_xlabel('x')
    ax[0, 1].set_ylabel('$\\Delta$u')
    ax[0, 1].legend()

    ax[1, 0].semilogy(deltaerr, lw=1, label='Maximum update')
    ax[1, 0].grid(linestyle=':')
    ax[1, 0].set_xlabel('Iteration')
    ax[1, 0].set_ylabel('$\\max(\\Delta$u)')
    ax[1, 0].legend()

    ax[1, 1].semilogy(residual, lw=1, label='Residual')
    ax[1, 1].grid(linestyle=':')
    ax[1, 1].set_xlabel('Iteration')
    ax[1, 1].set_ylabel('$\\mathrm{b-\\mathbf{A}\\cdot\,u^{\\prime}}$')
    ax[1, 1].legend()

    fig.tight_layout()
```

Main Jacobi iteration loop. At each step increase the iteration counter
that is used to ensure that the loop will not run forever in case of bad
convergence rate. Compute the new approximate solution that will be used
in next step. The convergence rate can be measured by the absolute change
between estimates `delta`. Another (and more formal) mean of doing so is
by computing the `residual` r = b - A u' after each iteration.

```python
def jacobi_iteration(size, L=1.0, relax=1.0, mxstep=5000,
                     eps=1.0e-06, verbose=False):
    """ Solves Poisson equation using Jacobi iteration.

    Parameters
    ----------
    size : int
        Number of discrete nodes in system.
    L : float
        Length of space domain.
    relax : float
        Under-relaxation factor.
    mxstep : int
        Maximum number of iterations.
    eps : float
        Residual convergence criterium.
    """
    
    # Make sure under-relaxation is in range [0, 1].
    relax = max(min(relax, 1.0), 0.0)
    
    x, dx2 = get_discretization(L, size)
    hrx, hrua, lrua = get_analytical_solution(x, L, size)
    A = get_problem_matrix(size, dx2, relax)

    # The right-hand side f(x)=1 of the problem.  Make sure that this variable
    # is not modified during solution, otherwise unexpected results will arise.
    b = np.ones(size)

    # Provide an initial `guess` for the problem, which will be updated across
    # iterations. Also provide the solution variable `u` that is used to update
    # the guess for next iteration.
    guess = np.zeros(size)
    u = guess.copy()

    # During solution, keep the absolute value of the maximum change in solution
    # array and the computed residual `r = b - Au`.
    deltaerr, residual = [], []

    counter = 0

    while True:
        guess, u, delta, resit = jacobi(A, b, u, relax, dx2, guess)
        deltaerr.append(delta)
        residual.append(resit)

        if resit < eps:
            if verbose:
                print("Converged after {:d} steps".format(counter))
            break

        if counter > mxstep:
            print("Unable to converge after {:d} steps".format(counter))
            break
    
        counter += 1
        
    plot_solution(x, u, hrx, hrua, lrua, deltaerr, residual, counter)
```

```python
%timeit jacobi_iteration(10, mxstep=5000)
```

```python
%timeit jacobi_iteration(10, relax=0.5, mxstep=5000)
```

```python
%timeit jacobi_iteration(10, relax=0.2, mxstep=5000)
```

```python
jacobi_iteration(100, mxstep=50000)
```
