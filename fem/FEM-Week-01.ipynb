# Solution to problems

## Problem 1

Consider the following differential equation of elastotatics. Find a satisfying

$$
\frac{d}{dx}\left(EA\frac{du}{dx}\right)+f=0\qquad\in\,(0,L)
$$

where $f$, $E$, and $A$ are constants, and s.t. the following set of boundary conditions are satisfied:

1. $u(0)=g_1$, $u(L)=g_2$
1. $EAu_x\vert_0=h_1$, $EAu_x\vert_L=h_2$, where $h_1=h_2+fL$.

---

### Solution of part 1

$$
\int_0^y\frac{d}{dx}\left(EA\frac{du}{dx}\right)+\int_0^yf=0\implies
EA\frac{du}{dx}\biggr\vert_0^y+fy=0
$$

which can be rearranged as

$$
\frac{du}{dy}=-\frac{f}{EA}y+\frac{du}{dy}\biggr\vert_0
$$

integrating once again now in terms of $z$ leads to

$$
u(z)-u(0)=-\frac{f}{2EA}z^2+\frac{du}{dy}\biggr\vert_0z
$$

replacing $u(0)=g_1$ produces

$$
u(z)=-\frac{f}{2EA}z^2+\frac{du}{dy}\biggr\vert_0z+g_1
$$

replacing $u(L)=g_2$ produces

$$
g_2=-\frac{f}{2EA}L^2+\frac{du}{dy}\biggr\vert_0L+g_1
$$

such that

$$
\frac{du}{dy}\biggr\vert_0=\frac{1}{L}\left(g_2-g_1+\frac{f}{2EA}L^2\right)
$$

and the general solution is

$$
u(z)=-\frac{f}{2EA}z^2+\left(\frac{g_2-g_1}{L}+\frac{f}{2EA}L\right)z+g_1
$$

---

### Solution of part 2

Starting from previous equation

$$
\frac{du}{dy}=-\frac{f}{EA}y+\frac{du}{dy}\biggr\vert_0
$$

with $EAu_x\vert_0=h_1$ one get

$$
\frac{du}{dy}=-\frac{f}{EA}y+\frac{h_1}{EA}
$$

and testing with $EAu_x\vert_L=h_2$

$$
\frac{h_2}{EA}=-\frac{f}{EA}L+\frac{h_1}{EA}\implies{}h_1=h_2+fL
$$

integrating once again now in terms of $z$ leads to

$$
u(z)=-\frac{f}{2EA}z^2+\frac{h_1}{EA}z+u(0)
$$

This solution is arbitrary up to the integration constant $u(0)$.

## Problem 2

For parts (a) through (c) consider the following boundary value problem in its strong and weak formulations:

(**S**) On the unit interval $\Omega=(0, 1)$ and $\bar{\Omega}=[0, 1]$, given $f:\Omega\rightarrow\Re$ and constants $A$, $B$, find $u:\bar{\Omega}\rightarrow\Re$ such that,

$$
\frac{d^4u}{dx^4}=f\quad\text{on}\quad\Omega
$$

$$
u(1)=0\quad\text{and}\quad\frac{du}{dx}\biggr\vert_1=0
$$

$$
\frac{d^2u}{dx^2}\biggr\vert_0=A\quad\text{and}\quad\frac{d^3u}{dx^3}\biggr\vert_0=B
$$

(**W**) $\mathcal{S}=\mathcal{V}=\{w\,\vert\,{}w\in{}H^2(\Omega), w(1)=w_x(1)=0\}$, given $f:\Omega\rightarrow\Re$ and constants $A$, $B$, find $u\in\mathcal{S}$ such that $\forall{}w\in\mathcal{V}$,

$$
\int_0^1w_{xx}u_{xx}dx=\int_0^1wfdx-w_x(0)A+w(0)B
$$

Also observe that $w\in{}H^2(\Omega)$ (i.e. $\int_{\Omega}\vert{w}\vert^2dx+\int_{\Omega}\vert{w_x}\vert^2dx+\int_{\Omega}\vert{w_{xx}}\vert^2dx<\infty$), unlike the problem in class where $w\in{}H^1(\Omega)$.

Prove the following propositions,

1. Let $u$ be a solution to (**S**), then $u$ is also a solution to (**W**).
1. Let $u$ be a solution to (**W**), then $u$ is also a solution to (**S**).

Also,

3. Would you call the boundary condition $u_x(1)=0$ as essential or natural? Give reasons.

---

### Proofs of 1-2

Multiply the strong form (**S**) by test function $w$ and integrate once by parts,

$$
\int_{\Omega}{}wu_{xxxx}dx=\int_{\Omega}wfdx\implies{}wu_{xxx}\big\vert_{\Omega}-\int_{\Omega}w_{x}u_{xxx}dx=\int_{\Omega}wfdx
$$

where the integration by parts $\int_{\Omega}udv=uv\vert_{\Omega}-\int_{\Omega}vdu$ was performed by making $u=w$ and $dv=u_{xxxx}dx$, such that $du=w_{x}dx$ and $v=u_{xxx}$. Applying B.C. $w(1)=0$ and $u_{xxx}(0)=B$ and leads to

$$
-\int_{\Omega}w_{x}u_{xxx}dx=\int_{\Omega}wfdx+w(0)B
$$

Repeating the integration by parts again and using the Dirichlet B.C. $w_x(1)=0$ and the Neumann B.C. $u_{xx}=A$ produces

$$
\int_{\Omega}w_{xx}u_{xx}dx=\int_{\Omega}wfdx-w_{x}(0)A+w(0)B
$$

which is actually the weak form (**W**). Starting with this expression and integrating twice by parts the other way around produces

$$
\int_{\Omega}wu_{xxxx}dx=\int_{\Omega}wfdx-w_{x}(0)A+w(0)B+w_{x}u_{xx}\big\vert_{\Omega}-wu_{xxx}\big\vert_{\Omega}
$$

Now since this holds $\forall{}w\in\mathcal{V}$, choose $w(x)=\phi(x)(u_{xxxx}-f)$, with $\phi(x)>0$ for $x\in\Omega$, $\phi(0)=\phi(1)=0$, and $\phi_x(0)=\phi_x(1)=0$. Substituting this in the previous equation leads to

$$
\int_{\Omega}\phi(x)(u_{xxxx} - f)^2dx=0\implies\phi(x)(u_{xxxx} - f)^2=0\quad\text{but}\quad\phi(x)\neq{}0
$$

thus the strong form is retrieved as

$$
u_{xxxx} = f,\quad\text{in}\,\Omega
$$

Now choose $w(x)=\psi(x)$ such that $\psi(0)=\psi(1)=0$, $\psi_x(0)\neq{}0$, and $\phi_x(1)=0$. Substituting this in the same equation and making use of strong form equation leads to $A-u_{xx}(0)=0\implies{}u_{xx}\vert_0=A$, as expected.

Next choose $w(x)=\chi(x)$ such that $\chi_x(0)\neq{}0$, $\chi(1)=0$, and $\chi_x(0)=\chi_x(1)=0$. Repeating the procedure leads to $B-u_{xxx}(0)=0\implies{}u_{xxx}\vert_0=B$, as expected.

### Comment on part 3

It is a natural boundary condition (Dirichlet) built-in to space $\mathcal{S}$.
